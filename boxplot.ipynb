```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dataset.csv")

df['durationMin'] = df['duration_ms'] / 60000

# duration buckets
bins = [0, 3, 4, 5, 10, df['durationMin'].max()]
labels = ['<3 min', '3-4 min', '4-5 min', '5-10 min', '10+ min']
df['DurationCategory'] = pd.cut(df['durationMin'], bins=bins, labels=labels, include_lowest=True)

# boxplot
plt.figure(figsize=(10,6))
sns.boxplot(x='DurationCategory', y='popularity', data=df)
plt.title('Popularity by Song Duration Category')
plt.xlabel('Duration Category')
plt.ylabel('Popularity')
plt.grid(True)
plt.tight_layout()
plt.show()
```
